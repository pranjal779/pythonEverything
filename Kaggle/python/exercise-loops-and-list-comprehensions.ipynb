On average, how much money can you expect to gain (or lose) every time you play the machine?  The casino keeps it a secret, but you can estimate the average value of each pull using a technique called the **Monte Carlo method**. To estimate the average outcome, we simulate the scenario many times, and return the average result.

Complete the following function to calculate the average value per play of the slot machine.

````python
def estimate_average_slot_payout(n_runs):
    """Run the slot machine n_runs times and return the average net profit per run.
    Example calls (note that return value is nondeterministic!):
    >>> estimate_average_slot_payout(1)
    -1
    >>> estimate_average_slot_payout(1)
    0.5
    """
    """Run the slot machine n_runs times and return the average net profit per run."""
    total_winnings = 0
    for _ in range(n_runs):  # Loop n_runs times
        total_winnings += play_slot_machine()  # Add each result to the total
    return total_winnings / n_runs  # Calculate the average
````

Here‚Äôs a breakdown of how to approach the problem step by step:

### Problem Context
- `play_slot_machine()` is a function that simulates a single spin of the slot machine and returns the winnings from that spin.
- You need to simulate playing the slot machine multiple times (e.g., 10 spins or 100 spins) and calculate the total winnings.

### Sample Problem Statement
**Write a function that simulates playing the slot machine `n` times and returns the total winnings.**  

---

### Steps to Solve

1. **Understand the function**: 
   - Make sure to understand the output of `play_slot_machine()`. Running it once will give you an idea of its behavior (e.g., mostly returning 0 but occasionally a larger number).

2. **Simulate multiple spins**: 
   - Use a loop to call `play_slot_machine()` multiple times.
   - Add the winnings from each spin to a total.

3. **Return the result**:
   - After the loop, return the total winnings.

---

### Solution with a Loop

Here‚Äôs a function to simulate playing the slot machine `n` times:

```python
def play_n_times(n):
    total_winnings = 0
    for _ in range(n):  # Repeat n times
        total_winnings += play_slot_machine()  # Add the winnings of each spin
    return total_winnings
```

---

### Explanation of Code
- **Initialization**: `total_winnings` starts at 0 to store the sum of winnings.
- **Loop**: The `for` loop iterates `n` times. Each iteration:
  - Calls `play_slot_machine()`.
  - Adds the result to `total_winnings`.
- **Return**: After the loop, the total is returned.

---

### Using List Comprehension

If you want to try it with list comprehension:

```python

    return sum(play_slot_machine() for _ in range(n))
```

Here:
- `play_slot_machine()` is called `n` times in a generator expression.
- `sum()` computes the total.

---

### Example Usage

If you want to play 10 times:

